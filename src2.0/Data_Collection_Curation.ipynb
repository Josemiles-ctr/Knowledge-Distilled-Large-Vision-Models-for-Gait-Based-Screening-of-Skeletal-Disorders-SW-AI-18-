# Gait Dataset Curation Summary
## Knowledge-Distilled LVM for Skeletal Disorder Screening

### Project Overview
This dataset supports the development of a large vision model for gait-based screening of skeletal disorders, specifically adapted for Ugandan populations and resource-limited healthcare contexts.

### Clinical Condition Distribution
- **Knee Osteoarthritis (KOA)**: 39.1% of dataset
- **Normal (Control)**: 23.4% of dataset  
- **Disabled**: 25.4% of dataset
- **Parkinson's Disease**: 12.1% of dataset

### Severity Stratification
The dataset includes clinically relevant severity grading:
- **KOA**: Early (30), Mild (40), Severe (30)
- **Parkinson's**: Early (14), Mild (12), Severe (5)
- **Disabled**: Assistive device usage (26) vs. Non-assistive (39)


### Technical Processing
- **Pose Estimation**: MediaPipe for skeletal landmark extraction
- **Output Format**: CSV files containing frame-by-frame landmark coordinates
- **Data Representation**: 256 CSV files (one per gait video)
- **Coordinate System**: 2D/3D skeletal keypoints with biomechanical relevance

### Key Biomechanical Features Extracted
- Joint angles and trajectories
- Stride length and cadence measures  
- Gait symmetry metrics
- Trunk sway and postural alignment
- Assistive device impact quantification
- etc.

### Dataset Strengths
1. **Clinical Relevance**: Severity-graded conditions enable progressive disease modeling
2. **Real-world Diversity**: Includes assistive device usage scenarios
3. **Control Group**: Substantial normal gait samples (n=60) for baseline comparison
4. **Multi-condition Coverage**: Addresses multiple skeletal disorders as outlined in project abstract
